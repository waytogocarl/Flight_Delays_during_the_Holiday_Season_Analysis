```python
#set dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy import stats
```

```python
# load aviation data
aviation_data = pd.read_csv("./dataset/airline_data.csv")
aviation_data.head()
```

```python
print(aviation_data["flight_date"].unique())
```

```python
flights = aviation_data.to_dict(orient='records')

filtered_flights = []

for flight in flights:
    flight_date = flight["flight_date"]
    if flight_date.startswith("11") or flight_date.startswith("12"):
        filtered_flights.append(flight)
```

```python
formatted_flights = []

random_flights = random.sample(filtered_flights, min(200, len(filtered_flights)))

for flight in random_flights:
    airline_name = flight["carrier_name"]
    departure_airport = flight["airport_name"]  # Remove the brackets to access the value
    departure_date = flight["flight_date"]  # Remove the brackets to access the value
    airline_delay = flight["carrier_delay"]  # Remove the brackets to access the value
    weather_delay = flight["weather_delay"]  # Remove the brackets to access the value
    nas_delay = flight["nas_delay"]  # Remove the brackets to access the value
    security_delay = flight["security_delay"]  # Remove the brackets to access the value
    late_aircraft_delay = flight["late_aircraft_delay"]  # Remove the brackets to access the value
    count_airline_delay = flight["carrier_ct"]  # Remove the brackets to access the value
    count_weather_delay = flight["weather_ct"]  # Remove the brackets to access the value
    count_nas_delay = flight["nas_ct"]  # Remove the brackets to access the value
    count_security_delay = flight["security_ct"]  # Remove the brackets to access the value
    count_late_aircraft_delay = flight["late_aircraft_ct"]  # Remove the brackets to access the value

    # NAS is the National Aviation System
    formatted_flights.append({
        "Airline Name": airline_name,
        "Departure Airport": departure_airport,
        "Departure Date": departure_date,
        "Delay due to Airline": airline_delay,
        "Delay due to Weather": weather_delay,
        "Delay due to NAS": nas_delay,
        "Delay due to Security": security_delay,
        "Delay due to Late Aircraft": late_aircraft_delay,
        "# of Airline Delays": count_airline_delay,
        "# of Weather Delays": count_weather_delay,
        "# of NAS Delays": count_nas_delay,
        "# of Security Delays": count_security_delay,
        "# of Late Aircraft Delays": count_late_aircraft_delay
    })
```

```python
flight_data_df = pd.DataFrame(formatted_flights)

print(flight_data_df)
```

```python
flight_data_df.to_csv("flights.csv", index=True, mode='w')

flight_data_df = pd.read_csv("flights.csv", index_col=0)

flight_data_df.head()
```
