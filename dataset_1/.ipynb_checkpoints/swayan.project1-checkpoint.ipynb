```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import seaborn as sns
from scipy import stats
```

```python
aviation_data = pd.read_csv("C:\\Users\\swaya\\Downloads\\Project_dataset\\dataset_1\\airline_data.csv")
aviation_data.head()
```

```python
print(aviation_data["flight_date"].unique())
```

```python
flights = aviation_data.to_dict(orient='records')

filtered_flights = []

for flight in flights:
    flight_date = str(flight["month"])
    if flight_date.startswith("11") or flight_date.startswith("12"):
        filtered_flights.append(flight)
        
print(len(filtered_flights))
```

```python
formatted_flights = []

random_flights = random.sample(filtered_flights, min(200, len(filtered_flights)))

print(f"Number of randomly selected flights: {len(random_flights)}")
```

```python
for flight in random_flights[:5]:
        print(flight["flight_date"])
```

```python
for flight in random_flights:
    airline_name = flight["carrier_name"]
    arrival_airport = flight["airport"]
    month = flight["month"]
    year = flight["year"]
    airline_delay = flight["carrier_delay"]
    weather_delay = flight["weather_delay"]
    nas_delay = flight["nas_delay"]
    security_delay = flight["security_delay"]
    late_aircraft_delay = flight["late_aircraft_delay"]
    count_airline_delay = flight["carrier_ct"]
    count_weather_delay = flight["weather_ct"]
    count_nas_delay = flight["nas_ct"]
    count_security_delay = flight["security_ct"]
    count_late_aircraft_delay = flight["late_aircraft_ct"]

    # NAS is the National Aviation System
    formatted_flights.append({
        "Year": year,
        "Month": month,
        "Airline Name": airline_name,
        "Arrival Airport": arrival_airport,
        "Delay due to Airline in Mins": airline_delay,
        "Delay due to Weather in Mins": weather_delay,
        "Delay due to NAS in Mins": nas_delay,
        "Delay due to Security in Mins": security_delay,
        "Delay due to Late Aircraft in Mins": late_aircraft_delay,
        "# of Airline Delays": count_airline_delay,
        "# of Weather Delays": count_weather_delay,
        "# of NAS Delays": count_nas_delay,
        "# of Security Delays": count_security_delay,
        "# of Late Aircraft Delays": count_late_aircraft_delay
    })
```

```python
flight_data_df = pd.DataFrame(formatted_flights)

print(flight_data_df)
```

```python
flight_data_df.to_csv("flights.csv", index=True, mode='w')

flight_data_df = pd.read_csv("flights.csv", index_col=0)

flight_data_df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(flight_data_df['Delay due to Airline in Mins'], bins=30, kde=True)
plt.title("Distribution of Flight Delays due to Airline")
plt.xlabel("Delay (minutes)")
plt.ylabel("Number of Flights")
plt.show()
```

```python
#2. Average Delay by Airline
#A bar plot showing the average delay time for each airline can reveal if certain airlines tend to have longer or shorter delays.
avg_delay_by_airline = flight_data_df.groupby('Airline Name')['# of Airline Delays'].mean().sort_values()

plt.figure(figsize=(12, 6))
avg_delay_by_airline.plot(kind='bar', color='skyblue')
plt.title("Average Delay by Airline")
plt.xlabel("Airline")
plt.ylabel("Number of Delay on Average")
plt.show()
```

```python
#2. Maximum Delay by Airline
#A bar plot showing the maximum delay time for each airline can reveal if certain airlines tend to have longer or shorter delays.
max_delay_by_airline = flight_data_df.groupby('Airline Name')['# of Airline Delays'].max().sort_values()

plt.figure(figsize=(12, 6))
max_delay_by_airline.plot(kind='bar', color='skyblue')
plt.title("MAX Delay by Airline")
plt.xlabel("Airline")
plt.ylabel("Number of Delay on Maximu")
plt.show()
```

```python
# Box Plot of Delay by Airline
#A box plot is helpful for examining the distribution and spread of delays by airline, giving insight into outliers and variability.

plt.figure(figsize=(12, 6))
sns.boxplot(data=flight_data_df, x='Airline Name', y='Delay due to Late Aircraft in Mins')
plt.title("Delay Distribution by Airline")
plt.xlabel("Airline")
plt.ylabel("Delay due to Late Aircraft in Mins")
plt.xticks(rotation=80)
plt.show()
```

```python
# Proportion of avg delay Flights by Airline
#A pie chart illustrating the proportion of avg delay flights by airline can be helpful if youâ€™re interested in each airlineâ€™s share in the dataset.

avg_delay_by_airline = flight_data_df.groupby('Airline Name')['# of Security Delays'].mean().sort_values()
plt.figure(figsize=(8, 8))
avg_delay_by_airline.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title("Proportion of Flights Avg Delay Airline")
plt.ylabel("")
plt.show()
```

```python
flight_data_df['Time_in_Hours'] = flight_data_df['Delay due to Airline in Mins'] / 60
print(flight_data_df['Time_in_Hours'])
```

```python
#Hourly Delay Trend Across the Day
#A line plot showing average delays across each hour of the day can reveal peak times with the most delays.

hourly_delays = flight_data_df.groupby('Time_in_Hours')['Delay due to Airline in Mins'].mean()

plt.figure(figsize=(12, 6))
hourly_delays.plot(kind='line', marker='o', color='purple')
plt.title("Average Delay by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Average Delay (minutes)")
plt.xticks(range(24))
plt.show()
```

```python
max_hourly_delays = flight_data_df.groupby('Time_in_Hours')['Delay due to Airline in Mins'].max()

plt.figure(figsize=(12, 6))
max_hourly_delays.plot(kind='line', marker='o', color='purple')
plt.title("max Delay by Hour of Day")
plt.xlabel("Time in Hours")
plt.ylabel("max Delay (minutes)")
plt.xticks(range(24))
plt.show()
```

```python
import matplotlib.pyplot as plt

# Manually define month names for each month number
month_names = {
     1:"Nvember", 2:"December"
}

# Map month numbers to month names in the DataFrame
 #flight_data_df['month'] = flight_data_df['month'].map(month_names)

#Group by 'Month Name' and 'Airline Name' to get the maximum delays
max_delay_by_airline_month = flight_data_df.groupby(['month', 'Airline Name'])['# of Airline Delays'].max()

# Loop through each month to generate pie charts for the top 5 airlines
for month, data in max_delay_by_airline_month.groupby(level=0):
    # Sort and select the top 5 airlines with the highest delays for the month
    top_5_delays = data.nlargest(5).droplevel(0)  # Remove month level for plotting
    
    # Plot pie chart
    plt.figure(figsize=(8, 8))
    top_5_delays.plot(kind='pie', autopct='%1.1f%%', startangle=140)
    plt.title(f"Top 5 Maximum Delay Flights by Airline for {"November","December"}")
    plt.ylabel("")  # Clear y-label for aesthetics
    plt.show()
```
